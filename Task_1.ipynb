```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv( '/content/housing_information.csv' )
df.dropna( inplace = True )
print( df.head() )
```

```python
x = df[ 'total_rooms' ].values
y = df[ 'median_house_value' ].values
x = x.reshape( len( x ), 1 )
y = y.reshape( len( y ), 1 )
```

```python
x_train = x[ : 14447 ] #First 14448 values will be the training data
x_test = x[ 14447 : ] #Rest values will be the testing data
y_train = y[ : 14447 ] #First 14448 values will be the training data
y_test = y[ 14447 : ] #Rest values will be the testing data
```

```python
regr_model = LinearRegression()
regr_model.fit( x_train, y_train )
plt.scatter( x_test, y_test, label = 'Original datapoints', color = 'red' )
plt.plot( x_test, regr_model.predict( x_test ), label = 'Predicted datapoints', color = 'black' )
plt.xlabel( 'Total number of rooms' )
plt.ylabel( 'Median house value' )
plt.legend( loc = 'upper right' )
plt.title( 'Plot of median house value w.r.t. total number of rooms' )
plt.show()
```
